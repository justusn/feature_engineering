````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("https://bit.ly/3deaKEM")
````

````python
df.columns = df.columns.str.replace(' ','_').str.replace('-','').str.replace(')','').str.replace('(','').str.replace('=','').str.replace('__','_').str.lower()
````

````python
# Drop unrequired columns
df = df.drop(['order_no',  'user_id'],axis = 1)
````

````python
# Too many 'precipitation_in_millimeters' nulls. Dropping the column
df = df.drop('precipitation_in_millimeters', axis=1) 
````

````python
# Adding speed
df["speed_kph"]=df["distance_km"]/(df["time_from_pickup_to_arrival"]/3600)
````

````python
# Adding manhattan distance
df["manhattan_distance"]= (df["pickup_lat"]-df["destination_lat"]) + (df["pickup_long"] - df["destination_long"])
````

````python
from datetime import timedelta

import datetime
````

````python
# Extract hour and minute
df['date'] = pd.to_datetime(df['placement_time'],format='%I:%M:%S %p')
df['placement_hour'] = pd.DatetimeIndex(df['date']).hour
df['placement_minute'] = pd.DatetimeIndex(df['date']).minute

df['date'] = pd.to_datetime(df['placement_time'],format='%I:%M:%S %p')
df['confirmation_hour'] = pd.DatetimeIndex(df['date']).hour
df['confirmation_minute'] = pd.DatetimeIndex(df['date']).minute

df['date'] = pd.to_datetime(df['arrival_at_pickup_time'],format='%I:%M:%S %p')
df['arrival_at_pickup_hour'] = pd.DatetimeIndex(df['date']).hour
df['arrival_at_pickup_minute'] = pd.DatetimeIndex(df['date']).minute

df['date'] = pd.to_datetime(df['arrival_at_pickup_time'],format='%I:%M:%S %p')
df['arrival_at_pickup_hour'] = pd.DatetimeIndex(df['date']).hour
df['arrival_at_pickup_minute'] = pd.DatetimeIndex(df['date']).minute

df['date'] = pd.to_datetime(df['pickup_time'],format='%I:%M:%S %p')
df['pickup_hour'] = pd.DatetimeIndex(df['date']).hour
df['pickup_minute'] = pd.DatetimeIndex(df['date']).minute

df['date'] = pd.to_datetime(df['arrival_at_destination_time'],format='%I:%M:%S %p')
df['arrive_destination_hour'] = pd.DatetimeIndex(df['date']).hour
df['arrive_destination_minute'] = pd.DatetimeIndex(df['date']).minute

````

````python
# Get time differentials
df["seconds"] = pd.to_datetime(df['confirmation_time'],format='%I:%M:%S %p') - pd.to_datetime(df['placement_time'],format='%I:%M:%S %p')
df["placement_to_confirmation_time"] = df["seconds"].dt.total_seconds()

df["seconds"] = pd.to_datetime(df['arrival_at_pickup_time'],format='%I:%M:%S %p') - pd.to_datetime(df['confirmation_time'],format='%I:%M:%S %p')
df["time_to_pickup_arrival"] = df["seconds"].dt.total_seconds()

df["seconds"] = pd.to_datetime(df['pickup_time'],format='%I:%M:%S %p') - pd.to_datetime(df['arrival_at_pickup_time'],format='%I:%M:%S %p')
df["time_to_pickup"] = df["seconds"].dt.total_seconds()

df["seconds"] = pd.to_datetime(df['arrival_at_destination_time'],format='%I:%M:%S %p') - pd.to_datetime(df['pickup_time'],format='%I:%M:%S %p')
df["time_to_deliver"] = df["seconds"].dt.total_seconds()

df["seconds"] = pd.to_datetime(df['arrival_at_destination_time'],format='%I:%M:%S %p') - pd.to_datetime(df['placement_time'],format='%I:%M:%S %p')
df["total_time_taken"] = df["seconds"].dt.total_seconds()
````

````python
# Drop timestamp columns
df = df.drop(["placement_time","confirmation_time","arrival_at_destination_time","arrival_at_pickup_time","pickup_time","date","seconds"],axis=1)
````
